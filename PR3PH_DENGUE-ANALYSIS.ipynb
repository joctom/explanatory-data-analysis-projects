Dengue is a viral infection caused by the dengue virus (DENV), transmitted to humans through the bite of infected mosquitoes. While many infected people show no symptoms, others experience high fever, severe headache, body aches, nausea, and a rash. This study aims to identify the trends of dengue outbreaks in the country during this five-year period.
