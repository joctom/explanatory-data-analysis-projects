<h3>Filling-Up Null Values</h3>

```python
penguins['bill_length_mm'].mean()
```

```python
penguins['bill_length_mm'] = penguins['bill_length_mm'].fillna(penguins['bill_length_mm'].mean())
penguins['bill_depth_mm'] = penguins['bill_depth_mm'].fillna(penguins['bill_depth_mm'].mean())
penguins['flipper_length_mm'] = penguins['flipper_length_mm'].fillna(penguins['flipper_length_mm'].mean())
penguins['body_mass_g'] = penguins['body_mass_g'].fillna(penguins['body_mass_g'].mean())

penguins.isnull().sum()
```

```python
penguins['sex'] = penguins['sex'].fillna(penguins['sex'].mode()[0])
penguins.isnull().sum()
```

```python
penguins['species'] = penguins['species'].astype('category')
penguins['island'] = penguins['island'].astype('category')
penguins['sex'] = penguins['sex'].astype('category')
penguins['year'] = penguins['year'].astype('category')
```

<h2> Question 8 </h2>

This size difference can be attributed to roles in breeding and competition, where larger body mass provides males with an advantage in securing mates and defending territories.

<h2> Question 9 </h2>
<h3> What percentage of each species' population is male? </h3>

```python
species_count = penguins['species'].value_counts()
male_count = penguins[penguins['sex'] == 'male'].groupby('species').size()
male_percentage = (male_count / species_count * 100).round(2)
male_percentage
```

<span style="color:red;"><b>INSIGHTS:</b></span> 

The male percentage is nearly a perfect 50/50 sex ratio across all three penguin species, with Adelie, Chinstrap, and Gentoo populations being 51.97%, 50.00%, and 53.23% male, respectively.

The numbers tell us the sex ratio is perfectly balanced across all species. This is a sign of a healthy, stable population. The very small differences between species are normal and not a cause for concern. In nature, a 50/50 split is common and suggests there are no major environmental pressures favoring one sex over the other.
<hr>

<h2> Question 10 </h2>
<h3> In what direction are bill length and bill depth related among Gentoo penguins? </h3>

```python
gentoo_corr = penguins[penguins['species'] == 'Gentoo']['bill_length_mm'].corr(penguins[penguins['species'] == 'Gentoo']['bill_depth_mm']).round(3)
print(f"Gentoo Bill Length/Depth Correlation: {gentoo_corr}")
```

<span style="color:red;"><b>INSIGHTS:</b></span> For Gentoo penguins, bill length and bill depth are positively correlated with a coefficient of 0.607. This is a strong relationship, meaning Gentoo penguins with longer bills also tend to have deeper bills. While the association is not perfect, it is consistent enough that bill length can be considered a fairly good predictor of bill depth within this species.
<hr>

<h2> Question 11 </h2>
<h3> Determine the central flipper length measurement for male and female Chinstrap penguins on Dream Island. </h3>

```python
chinstrap_flipper = penguins[(penguins['species'] == 'Chinstrap') & (penguins['island'] == 'Dream')].groupby('sex')['flipper_length_mm'].median()
chinstrap_flipper
```

<span style="color:red;"><b>INSIGHTS:</b></span> This illustrates the difference between male and female Chinstrap penguins on Dream Island. The median flipper length for females is 192.0 mm, while for males it is 200.5 mm. 

This 8.5 mm difference shows that male Chinstrap penguins are consistently larger than females, a common biological pattern known as sexual dimorphism.

<h2> Question 12 </h2>
<h3> Show the distribution of penguin species for each study year. </h3>

```python
year_species = penguins.groupby(['year', 'species']).size()
print(year_species)
```

The Adelie population remained very stable each year, with around 50 individuals. The Gentoo population displayed a positive trend, increasing from 34 to 44 individuals. In contrast, the Chinstrap population was more variable, dropping to 18 individuals in 2008 before recovering in 2009.

This may mean that changes in the environment or in how the penguins were observed affected the Chinstrap population during the study period.

<h2> Question 13 </h2>
<h3> Does body mass relate more strongly to flipper length or to bill length in penguins? </h3>

```python
corr_mass_flipper = penguins['body_mass_g'].corr(penguins['flipper_length_mm']).round(3)
corr_mass_bill = penguins['body_mass_g'].corr(penguins['bill_length_mm']).round(3)
print(f"Correlation (Mass vs. Flipper): {corr_mass_flipper}")
print(f"Correlation (Mass vs. Bill Length): {corr_mass_bill}")
```

<span style="color:red;"><b>INSIGHTS:</b></span> The correlation between body mass and flipper length is 0.871, indicating a very strong positive relationship. This means that as penguinsâ€™ body mass increases, their flipper length also tends to increase in a highly consistent way. In contrast, the correlation between body mass and bill length is 0.595, which represents a moderate positive relationship. Heavier penguins generally have longer bills, but the connection is not as strong.
